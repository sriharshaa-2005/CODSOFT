TEXT VECTORIZATION USING TF-IDF:

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    stop_words='english'
)
```

```python
X = tfidf.fit_transform(sd['v2'])
y = sd['v1']
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

SPLITTING,TRAINING AND TESTING OF DATA:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

IMPLEMENTATION OF LOGISTIC REGRESSION MODEL:

```python
lr= LogisticRegression(max_iter=1000, class_weight='balanced')
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

PREDICTION FOR THE NEW DATA USING LOGISTIC REGRESSION MODEL:

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)
```

```python
new_messages = [
    "Congratulations! You’ve won a free ticket to Bahamas. Call now!",
    "Hey, are we still on for lunch tomorrow?",
    "URGENT! Your account has been compromised. Click here to reset your password.",
    "Hi mom, just letting you know I got home safely. Love you!",
    "Free entry to our $5000 prize draw. Text 'WIN' to 12345 to enter now!",
    "Please find the attached invoice for your recent purchase.",
    "Important update: Your subscription will expire soon. Renew now!"
]
```

```python
new_messages_tfidf = tfidf.transform(new_messages)
predictions = lr.predict(new_messages_tfidf)
for msg, label in zip(new_messages, predictions):
    print(f"Message: {msg}\nPredicted Category: {'Spam' if label == 1 else 'Ham'}\n")
```
